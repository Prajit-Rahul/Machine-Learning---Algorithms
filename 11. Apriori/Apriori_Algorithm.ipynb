```python
from apyori import apriori
```

```python
dataset = pd.read_csv('dataset1.csv')
```

```python
print(dataset.shape)
dataset.head(10)
```

```python
transactions = []

for i in range(0, 7500):

  transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])

print(transactions)
```

```python
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 3, max_length = 3)
```

```python
results = list(rules)
print(results)
```

```python
for i in results:
  print(i)
```

```python
lhs         = [tuple(result[2][0][0])[0] for result in results]
rhs         = [tuple(result[2][0][1])[0] for result in results]
supports    = [result[1] for result in results]
confidences = [result[2][0][2] for result in results]
lifts       = [result[2][0][3] for result in results]

resultsinDataFrame = pd.DataFrame(zip(lhs, rhs, supports, confidences, lifts), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])

resultsinDataFrame
```
