```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('dataset.csv')
```

```python
import scipy.cluster.hierarchy as clus
```

```python
plt.figure(1,figsize=(16,81))
dentogram = clus.dendrogram(clus.linkage(dataset,method = 'ward'))
plt.title('Dendogram Tree Graph')
plt.xlabel('INCOME')
plt.ylabel('SPENT')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
model = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='average')
y_means = model.fit_predict(dataset)
y_means
```

```python
X = dataset.iloc[:,:].values
X
```

```python
plt.scatter(X[y_means == 0,0], X[y_means == 0,1],s=50,c='purple',label='cluster1')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1], s= 50, c = 'blue', label = 'cluster2')
plt.scatter(X[y_means == 2,0], X[y_means == 2,1],s=50,c='green',label='cluster3')
plt.title('INCOME-SPENT ANALYSIS HIERARCHIAL CLUSTERING')
plt.xlabel('INCOME')
plt.ylabel('SPENT')
plt.legend()
plt.show()
```
